```python
# m03 - Functional vs OOP Programming
# author: WJL
# created: 2022-02-5
# Program that uses classes to create a list of animals that an employee checks in at the vet


# ADDED LINE BREAKS TO MAKE CODE EASIER TO READ

# Creating the "Pet" class
class Pet:
    def __init__(self, pet_type, name, age, weight, color, is_vaccinated, note):
        self.pet_type = pet_type
        self.name = name
        self.age = age
        self.weight = weight
        self.color = color
        self.is_vaccinated = is_vaccinated
        self.note = note
    
    def __repr__(self):
        return f" Pet type: {self.pet_type}\n Pet name: {self.name}\n Pet age: {self.age}\n Pet weight: {self.weight}\n\
 Pet color: {self.color}\n Is pet vaccinated? {self.is_vaccinated}\n Notes: {self.note}"
        
# Creating "Dog" class that inherits from parent "Pet" class
class Dog(Pet):
    def __init__(self, pet_type, name, age, weight, color, is_vaccinated, is_neutered, is_spayed, temperment, note):
        self.is_neutered = is_neutered
        self.is_spayed = is_spayed
        self.temperment = temperment
        super().__init__(pet_type, name, age, weight, color, is_vaccinated, note)
        
    def __repr__(self):
        return f"Pet type: {self.pet_type}\n Pet name: {self.name}\n Pet age: {self.age}\n Pet weight: {self.weight}\n\
 Pet color: {self.color}\n Is pet vaccinated? {self.is_vaccinated}\n Is pet neutered? {self.is_neutered}\n\
 Is pet spayed? {self.is_spayed}\n Pet temperment: {self.temperment}\n Notes: {self.note}"

# Creating "Cat" class that inherits from parent "Pet" class
class Cat(Pet):
    def __init__(self, pet_type, name, age, weight, color, is_vaccinated, is_neutered, is_spayed, allergies, note):
        self.is_neutered = is_neutered
        self.is_spayed = is_spayed
        self.allergies = allergies
        super().__init__(pet_type, name, age, weight, color, is_vaccinated, note)
        
    def __repr__(self):
        return f"Pet type: {self.pet_type}\n Pet name: {self.name}\n Pet age: {self.age}\n Pet weight: {self.weight}\n\
 Pet color: {self.color}\n Is pet vaccinated? {self.is_vaccinated}\n Is pet neutered? {self.is_neutered}\n\
 Is pet spayed? {self.is_spayed}\n Pet's allergies: {self.allergies}\n Notes: {self.note}"

# Creating each object and assigning them their attributes
Teddy = Dog("Dog", "Teddy", 14, 10, "Brown and black", "No", "Yes", "No", "Good", "Teddy is a yorkie, although he is a very\
 good tempered dog he gets mad if another animal steals his toy")

Max = Dog("Dog", "Max", 12, 60, "Golden", "Yes", "Yes", "No", "Playful", "Max is a goldendoodle, he is very playful and enjoys\
 running.")

Ralf = Dog("Dog", "Ralf", 2, 8, "white", "No", "Yes", "No", "Reserved", "Ralf is a toy poodle, while he isn't the most outgoing\
 little pup he is very kind.")

Mr_Whiskers = Cat("Cat", "Mr. Whiskers", 4, 9, "Black", "Yes", "Yes", "No", "Tuna", "Mr. Whiskers is a very fun loving cat who\
 enjoys being around others. IMPORTANT: Do not feed him Tuna")

Hershey = Cat("Cat", "Hershey", 3, 6, "Black", "No", "No", "Yes", "None", "Hershey is a mischievous young cat who is very nosey")

Fred = Cat("Cat", "Fred", 14, 10, "Tabby", "Yes", "Yes", "No", "None", "Fred is very outgoing and good at making new friends,\
 he loves attention")

# Printing the animals
print("Dogs:\n\nTeddy:\n", Teddy)
print("\nMax:\n", Max)
print("\nRalf:\n", Ralf)
print("\nCats:\n\nMr. Whiskers:\n", Mr_Whiskers)
print("\nHershey:\n", Hershey)
print("\nFred:\n", Fred)
```
